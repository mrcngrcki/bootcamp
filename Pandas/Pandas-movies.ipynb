```python
import pandas as pd
import numpy as np
```

```python
colnames = ['MovieID', 'Title', 'Genres']
movies = pd.read_csv('ml-1m/movies.dat',
                     sep='::',
                     engine='python',
                     header=None,
                     names=colnames)
#movies['Year'] = movies['Title'].str.extract('\((\d{4})\)',expand=False)
#movies['Title'].str.extract('\((\d{4})\)',expand=False)
#movies.Title.str.slice(-5,-1)


movies['Year'] = movies.Title.str[-5:-1].astype(int)
movies['Title'] = movies['Title'].str[:-6].str.strip()
movies
```

```python
colnames = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
users = pd.read_csv('ml-1m/users.dat', sep='::', engine='python', header=None, names=colnames)
users
```

```python
colnames = ['UserID', 'MovieID', 'Rating', 'Timestamp']
ratings = pd.read_csv('ml-1m/ratings.dat', sep='::', engine='python', header=None, names=colnames)
ratings.head(20)
```

## Wykryj i wypisz wszystkie remake'i - ten sam tytul, rozne lata

```python
counts = movies.Title.value_counts()
counts[counts > 1].reset_index().rename({'Title':'Counts', 'index':'Title'})
```

```python
movies_duplicated = movies[movies.duplicated(subset='Title', keep=False)]
movies_year = movies_duplicated.groupby('Title')['Year'].agg(['min', 'max'])
movies_year
movies_year['Diff'] = movies_year['max'] - movies_year['min']
movies_year.sort_values(by='Diff', ascending=False)
#movies_year['min']
```

## najlepszy film wszechczasow

```python
mrating = ratings.groupby('MovieID').agg({'Rating':'mean'}).sort_values(by='Rating', ascending=False)
mrating.merge(movies, on='MovieID')
```

## to samo ale liczba ocen > 100

```python
mrating = ratings.groupby('MovieID')['Rating'].agg({'Rating':'mean', 'Size':'size'}).sort_values(by='Rating', ascending=False)
#mrating.merge(movies, on='MovieID')
mrating[mrating['Size'] > 100].sort_values(by='Rating', ascending=False)
```

## jaki jest najlepszy wg kobiet i wg mezczyzn

```python
mrating = ratings.merge(users).groupby(['MovieID', 'Gender'])['Rating'].mean().sort_values(ascending=False)
mrating.reset_index()
```
