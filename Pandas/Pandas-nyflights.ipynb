# Zadanie 1

```python
airlines = pd.read_csv('nycflights13_airlines.csv.gz', compression='gzip', comment='#')
airports = pd.read_csv('nycflights13_airports.csv.gz', compression='gzip', comment='#')
flights = pd.read_csv('nycflights13_flights.csv.gz', compression='gzip', comment='#')
planes = pd.read_csv('nycflights13_planes.csv.gz', compression='gzip', comment='#')
weather = pd.read_csv('nycflights13_weather.csv.gz', compression='gzip', comment='#')


con = sqlite3.connect('nycflights13.db')
planes.to_sql("planes", con, if_exists="replace")
airlines.to_sql("airlines", con, if_exists="replace")
flights.to_sql("flights", con, if_exists="replace")
airports.to_sql("airports", con, if_exists="replace")
weather.to_sql("weather", con, if_exists="replace")

sql_result = pd.read_sql_query("SELECT * FROM planes", con)
sql_result.head()
```

## SELECT DISTINCT engine FROM planes

```python
sql_result = pd.read_sql_query("SELECT DISTINCT engine FROM planes", con)
sql_result

```

```python
planes_unique = pd.DataFrame({"engine":pd.unique(planes['engine'])})
planes_unique
```

## SELECT DISTINCT type, manufacturer FROM planes

```python
sql_result = pd.read_sql_query("SELECT DISTINCT type, manufacturer FROM planes", con)
sql_result
```

```python
#planes.drop_duplicates(subset=['type','manufacturer'], keep='first')[['type','manufacturer']].reset_index()

planes[['type','manufacturer']].drop_duplicates().reset_index(drop=True)
```

## SELECT COUNT(*), engine FROM planes GROUP BY engine

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result
```

```python
pd.DataFrame(planes.groupby('engine').size()).reset_index().rename(columns={0:'COUNT(*)'})\
.reindex(columns=['COUNT(*)', 'engine'])
```

## SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type", con)
sql_result
```

```python
pd.DataFrame(planes.groupby(['engine','type']).size()).reset_index()\
.rename(columns={0:'COUNT(*)'})[['COUNT(*)','engine','type']]
```

## SELECT tailnum FROM planes WHERE year >= 2010

```python
sql_result = pd.read_sql_query\
("SELECT tailnum FROM planes WHERE year >= 2010", con)
sql_result
```

## SELECT tailnum FROM planes WHERE seats BETWEEN 100 and 200 LIMIT 20

```python
sql_result = pd.read_sql_query\
("SELECT tailnum FROM planes WHERE seats BETWEEN 100 and 200 LIMIT 20", con)
sql_result
```

```python
pd.DataFrame(planes[planes.seats.between(100,200)]['tailnum']).reset_index(drop=True).head(20)
```

## SELECT * FROM planes WHERE manufacturer IN ("BOEING", "AIRBUS", "EMBRAER")

```python
sql_result = pd.read_sql_query\
("SELECT * FROM planes WHERE manufacturer IN ('BOEING', 'AIRBUS', 'EMBRAER')", con)
sql_result
```

```python
pd.DataFrame(planes[planes.manufacturer.isin(['BOEING','AIRBUS','EMBRAER'])]).reset_index()
```

## SELECT * FROM planes WHERE manufacturer IN ("BOEING", "AIRBUS", "EMBRAER") AND seats>300

```python
sql_result = pd.read_sql_query\
("SELECT * FROM planes WHERE manufacturer IN ('BOEING', 'AIRBUS', 'EMBRAER') AND seats>300", con)
sql_result
```

```python
pd.DataFrame(planes[planes.manufacturer.isin(['BOEING','AIRBUS','EMBRAER']) & (planes.seats>300)])\
.reset_index()
```

## SELECT manufacturer, COUNT(*) FROM planes WHERE seats > 200 GROUP BY manufacturer

```python
sql_result = pd.read_sql_query\
("SELECT manufacturer, COUNT(*) FROM planes WHERE seats > 200 GROUP BY manufacturer", con)
sql_result
```

```python
pd.DataFrame(planes[planes.seats>200].groupby(['manufacturer']).size()).reset_index().rename(columns={0:'COUNT(*)'})
```

## SELECT manufacturer, COUNT(*) FROM planes GROUP BY manufacturer HAVING COUNT(*) > 10

```python
sql_result = pd.read_sql_query\
("SELECT manufacturer, COUNT() FROM planes GROUP BY manufacturer HAVING COUNT() > 10", con)
sql_result
```

```python
pd.DataFrame(planes.groupby(['manufacturer']).size()).reset_index().rename(columns={0:'COUNT()'})\
.filter(lambda x: len(x) > 10)
```

## SELECT manufacturer, COUNT(\*) FROM planes WHERE seats > 200 GROUP BY manufacturer HAVING COUNT(\*) > 10

```python
sql_result = pd.read_sql_query\
("SELECT manufacturer, COUNT(*) FROM planes WHERE seats > 200 GROUP BY manufacturer HAVING COUNT(*) > 10", con)
sql_result
```

```python
wynik = planes[planes.seats>200].groupby('manufacturer').size()
pd.DataFrame(wynik[wynik>10]).reset_index()
```

## SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany

```python
sql_result = pd.read_sql_query\
("SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany", con)
sql_result
```

```python
planes.groupby('manufacturer').size().reset_index().rename(columns={0:'howmany'}).sort_values('howmany')
```

## SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany DESC LIMIT 10

```python
planes.groupby('manufacturer').size().reset_index().rename(columns={0:'howmany'}).sort_values('howmany', ascending=False).head(10)
```

## SELECT \* FROM planes WHERE year >= 2012 ORDER BY year, seats

```python
sql_result = pd.read_sql_query\
("SELECT * FROM planes WHERE year >= 2012 ORDER BY year, seats", con)
sql_result
```

```python
planes[planes.year >= 2012].sort_values(['year','seats']).reset_index()
```
