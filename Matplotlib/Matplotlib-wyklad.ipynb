## Kanał alpha

```python
# system szesnastkowy ma cyfry od 0 do F
# https://www.rapidtables.com/web/color/RGB_Color.html
t = np.linspace(0, 2*np.pi, 100, endpoint=False)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.fill(x-2, y, fill=True, color = "#FF00AA") # RRGGBB
plt.fill(x, y, fill=True, color="#FF0000")
plt.fill(x+2, y, fill=True, color="#00FF00", alpha=0.4) # 0.0 to pełna przezroczystość, 1.0 to pełna nieprzezroczystość
plt.show()
```

```python
# system szesnastkowy ma cyfry od 0 do F
# https://www.rapidtables.com/web/color/RGB_Color.html
t = np.linspace(0, 2*np.pi, 100, endpoint=False)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.fill(x-2, y, fill=True, color = "#FF00AAF0") # RRGGBB
plt.fill(x, y, fill=True, color="#FF00002F")
plt.fill(x+2, y, fill=True, color="#00FF00E5") # 0.0 to pełna przezroczystość, 1.0 to pełna nieprzezroczystość
plt.show()
```

# Histogram

```python
import matplotlib.mlab as mlab
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x,
                            30,
                            density=False,
                            facecolor='green',
                            alpha=0.75,
                            edgecolor="k")

# y = mlab.normpdf(bins, mu, sigma)
# l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
# plt.axis([40, 160, 0, 0.03])
plt.grid(False)
plt.show()
```
