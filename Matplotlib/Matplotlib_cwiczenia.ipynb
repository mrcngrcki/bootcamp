```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
```

```python
def kolo(x0=0, y0=0, r=2, facecolor='#FF6200', edgecolor='#0000CC', n=100):

    t = np.linspace(0, 2*np.pi, n, endpoint=False)
    x = x0 + r*np.cos(t)
    y = y0 + r*np.sin(t)
    plt.fill(x, y, fill=True, facecolor=facecolor, edgecolor = edgecolor)
    plt.axis("equal")
    plt.plot(0,0,'ko')
    plt.text(0,0,'(0,0)', horizontalalignment = 'left', verticalalignment = 'top')
    
    
kolo()
kolo(3,4,2)
kolo(1,1,3)
plt.show()
```

```python
import seaborn as sns
flights = sns.load_dataset("flights")
sredniPasazerowie = flights["passengers"].groupby(flights["year"]).agg(np.mean)
sredniPasazerowie
```

```python
plt.plot(sredniPasazerowie, "bd-.")
plt.grid()
plt.legend(['Srednia liczba pasazerow'])
plt.tick_params(axis='x', colors='red')
plt.tick_params(axis='y', colors='blue')
plt.title = 'Srednia miesieczna liczba pasazerow w kolejnych latach'
```
