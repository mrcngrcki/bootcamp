## 1 Wyświetl tytuły albumów dostępnych w sklepie w kolejności alfabetycznej.

```python
c.execute('SELECT Title FROM albums ORDER BY Title')
pprint(c.fetchall())
```

## 2 Wyświetl imię, nazwisko i tytuł pracownika. Listę posortuj w malejąco po nazwisku.

```python
c.execute('SELECT FirstName, LastName FROM employees ORDER BY LastName DESC')
pprint(c.fetchall())
```

## 3 Ile albumów jest dostępnych w sklepie?

```python
c.execute('SELECT COUNT(*) FROM albums')
pprint(c.fetchall())
```

## 4 Jaka jest średnia długość utworu sprzedawanego w sklepie?

```python
c.execute('SELECT AVG(Milliseconds) FROM tracks')
pprint(c.fetchall())
```

## 5 Ile trwa najkrótszy utwór w dostępny w sklepie sklepie?

```python
c.execute('SELECT MIN(Milliseconds) FROM tracks')
pprint(c.fetchall())
```

## 6 Utwórów którego artysty jest najwięcej w sklepie?

```python
c.execute('SELECT Count(*) as max_artist, a.ArtistId, art.Name FROM Albums as a join Artists as art on a.ArtistId = art.ArtistId GROUP BY art.ArtistId ORDER BY max_artist desc')
pprint(c.fetchall())
```

## 8 Który pracownik opiekuje się największą liczbą klientów?

```python
c.execute('SELECT Count(*) as max_employee, c.SupportRepId, e.FirstName, e.LastName FROM Customers as c join Employees as e on c.SupportRepId = e.EmployeeId GROUP BY c.SupportRepId ORDER BY max_employee desc')
pprint(c.fetchall())
```

## 9 Ile wynosi najwyższa kwota na fakturze?

```python
c.execute('SELECT Max(Total) from invoices')
pprint(c.fetchall())
```

## 10 Który klient wydał w sklepie najmniej pieniędzy?

```python
c.execute('SELECT Min(Total), i.CustomerId, c.FirstName, c.LastName from invoices as i join customers as c on i.CustomerId = c.CustomerId')
pprint(c.fetchall())
```
