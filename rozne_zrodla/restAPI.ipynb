# Pobrac dane z grudnia 2019, wrzucic do dataframe i policzyc srednia euro i dolara z tego czasu

```python
import pandas as pd

startDate = '2019-12-01'
endDate = '2019-12-31'
table = 'A'

myList = []
    
url = f'http://api.nbp.pl/api/exchangerates/tables/{table}/{startDate}/{endDate}/'
params = {'format': 'json'}
resp = requests.get(url, params=params)
response = resp.json()
for table in response:
    date = table['effectiveDate']
    for rate in table['rates']:
        currency_name = rate['currency']
        currency_code = rate['code']
        currency_mid = rate['mid']
        myList.append((date, currency_name, currency_code, currency_mid))
        
df = pd.DataFrame(myList, columns=['Data', 'Waluta', 'Kod', 'Kurs'])
df
```

```python
print(f"Srednia Euro: {df[df['Kod']=='EUR'].agg({'Kurs':'mean'})}")
print(f"Srednia USD: {df[df['Kod']=='USD'].agg({'Kurs':'mean'})}")
```
